Index store notebook: define `IndexEntry` dataclass and the event-loop-only in-memory `index` dict.

```python
from __future__ import annotations

from dataclasses import dataclass


@dataclass(slots=True)
class IndexEntry:
    url: str
    origin_url: str
    depth: int
    title: str
    body_text: str
    crawled_at: float  # Unix timestamp

```
